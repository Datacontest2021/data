```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

elec_car= pd.read_csv('C:\\신현호\\코딩스터디\\데이터_코드\\2018년_전기자동차_월별.csv', encoding = 'utf-8')
elec_car
```

```python
#서울 행 추출
data_s = elec_car.loc[0]
data_s = data_s.values.reshape(1,-1)
data_s = data_s.tolist()
data_s = sum(data_s, [])
data_s = data_s[1:]

#경기 행 추출
data_g = elec_car.loc[1]
data_g = data_g.values.reshape(1,-1)
data_g = data_g.tolist()
data_g = sum(data_g, [])
data_g = data_g[1:]

#인천 행 추출
data_i = elec_car.loc[2]
data_i = data_i.values.reshape(1,-1)
data_i = data_i.tolist()
data_i = sum(data_i, [])
data_i = data_i[1:]

#대구 행 추출
data_d = elec_car.loc[3]
data_d = data_d.values.reshape(1,-1)
data_d = data_d.tolist()
data_d = sum(data_d, [])
data_d = data_d[1:]

#제주 행 추출
data_j = elec_car.loc[4]
data_j = data_j.values.reshape(1,-1)
data_j = data_j.tolist()
data_j = sum(data_j, [])
data_j = data_j[1:]
```

```python
plt.rc('font', family='NanumGothic')

fig = plt.figure(figsize=(10,10))
fig.set_facecolor('white')
ax = fig.add_subplot()

line = ['1월','2월','3월','4월','5월','6월','7월','8월','9월','10월','11월','12월']
ax.plot(line,data_s,marker='o',label='서울')
ax.plot(line,data_g,marker='o',label='경기')
ax.plot(line,data_i,marker='o',label='인천')
ax.plot(line,data_d,marker='o',label='대구')
ax.plot(line,data_j,marker='o',label='제주')

ax.legend()
plt.title('월별 전기자동차 수',fontsize=20)
plt.show()
```
