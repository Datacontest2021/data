```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

co_01= pd.read_csv('C:\\신현호\\코딩스터디\\CO오염도_1월\\18년도_1월_CO오염도.csv', encoding = 'utf-8')

co_01 = co_01.drop(columns = ['측정소명'])
#co_01 = co_01.reset_index()
#co_01 = co_01.drop(columns = ['index'])
co_01
```

```python
#히스토그램
import matplotlib.font_manager as fm

list_01 = []
list_02 = []
data_01 = co_01['시군구'].values.reshape(1,-1)
data_02 = co_01['오염도 월평균(ppm)'].values.reshape(1,-1)

data_01 = data_01.tolist()
data_02 = data_02.tolist()
for i in data_01:
    list_01.append(i)
for j in data_02:
    list_02.append(j)

#sum(my_list, [])
list_01 = sum(list_01,[])
list_02 = sum(list_02,[])

# font_path = r'C:\\Users\\hhshi\\Downloads\\NanumFontSetup_TTF_ALL\\NanumBarunGothic.ttf'
# fontprop = fm.FontProperties(fname=font_path, size=10)

plt.rc('font', family='NanumGothic')
#print(plt.rcParams['font.family'])

plt.title('2018년도 1월 CO 오염도')
plt.bar(list_01, list_02, color = 'b',width = 0.4)
plt.xlabel('시군구')
plt.ylabel('CO 오염도')
plt.xticks(rotation='vertical')
plt.legend(labels= ['오염도 수치'], loc='best', fontsize = 10)
plt.show()
```
