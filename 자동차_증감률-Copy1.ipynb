```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

elec_car = pd.read_csv('C:\신현호\코딩스터디\data_usage\전기자동차_최종.csv', encoding = 'utf-8')
norm_car = pd.read_csv('C:\신현호\코딩스터디\data_usage\일반자동차_최종.csv', encoding = 'utf-8')
co_accuracy = pd.read_csv('C:\신현호\코딩스터디\data_usage\CO오염도_최종.csv', encoding = 'utf-8')
elec_car
```

```python
#전기자동차
#서울 행 추출
data_s = elec_car.loc[0]
data_s = data_s.values.reshape(1,-1)
data_s = data_s.tolist()
data_s = sum(data_s, [])
data_s = data_s[1:]

#경기 행 추출
data_g = elec_car.loc[1]
data_g = data_g.values.reshape(1,-1)
data_g = data_g.tolist()
data_g = sum(data_g, [])
data_g = data_g[1:]

#인천 행 추출
data_i = elec_car.loc[2]
data_i = data_i.values.reshape(1,-1)
data_i = data_i.tolist()
data_i = sum(data_i, [])
data_i = data_i[1:]

#대구 행 추출
data_d = elec_car.loc[3]
data_d = data_d.values.reshape(1,-1)
data_d = data_d.tolist()
data_d = sum(data_d, [])
data_d = data_d[1:]

#제주 행 추출
data_j = elec_car.loc[4]
data_j = data_j.values.reshape(1,-1)
data_j = data_j.tolist()
data_j = sum(data_j, [])
data_j = data_j[1:]
```

```python
#일반자동차
#서울 행 추출
data_ns = norm_car.loc[0]
data_ns = data_ns.values.reshape(1,-1)
data_ns = data_ns.tolist()
data_ns = sum(data_ns, [])
data_ns = data_ns[1:]

#경기 행 추출
data_ng = norm_car.loc[1]
data_ng = data_ng.values.reshape(1,-1)
data_ng = data_ng.tolist()
data_ng = sum(data_ng, [])
data_ng = data_ng[1:]

#인천 행 추출
data_ni = norm_car.loc[2]
data_ni = data_ni.values.reshape(1,-1)
data_ni = data_ni.tolist()
data_ni = sum(data_ni, [])
data_ni = data_ni[1:]

#대구 행 추출
data_nd = norm_car.loc[3]
data_nd = data_nd.values.reshape(1,-1)
data_nd = data_nd.tolist()
data_nd = sum(data_nd, [])
data_nd = data_nd[1:]

#제주 행 추출
data_nj = norm_car.loc[4]
data_nj = data_nj.values.reshape(1,-1)
data_nj = data_nj.tolist()
data_nj = sum(data_nj, [])
data_nj = data_nj[1:]
```

```python
#전기자동차 증감률 계산
#서울
up_down_elec_s = []

for i in range(len(data_s)-1):
    up_down_elec_s.append(round((data_s[i+1]-data_s[i])/data_s[i]*100,2))
    
#up_down_elec_s
#경기
up_down_elec_g = []

for i in range(len(data_g)-1):
    up_down_elec_g.append(round((data_g[i+1]-data_g[i])/data_g[i]*100,2))
#인천
up_down_elec_i = []

for i in range(len(data_i)-1):
    up_down_elec_i.append(round((data_i[i+1]-data_i[i])/data_i[i]*100,2))
#대구
up_down_elec_d = []

for i in range(len(data_d)-1):
    up_down_elec_d.append(round((data_d[i+1]-data_d[i])/data_d[i]*100,2))
#제주
up_down_elec_j = []

for i in range(len(data_j)-1):
    up_down_elec_j.append(round((data_j[i+1]-data_j[i])/data_j[i]*100,2))
up_down_elec_j
```

```python
#일반자동차 증감률 계산
#서울
up_down_norm_s = []

for i in range(len(data_ns)-1):
    up_down_norm_s.append(round((data_ns[i+1]-data_ns[i])/data_ns[i]*100,2))
    
#up_down_elec_s
#경기
up_down_norm_g = []

for i in range(len(data_ng)-1):
    up_down_norm_g.append(round((data_ng[i+1]-data_ng[i])/data_ng[i]*100,2))
#인천
up_down_norm_i = []

for i in range(len(data_ni)-1):
    up_down_norm_i.append(round((data_ni[i+1]-data_ni[i])/data_ni[i]*100,2))
#대구
up_down_norm_d = []

for i in range(len(data_nd)-1):
    up_down_norm_d.append(round((data_nd[i+1]-data_nd[i])/data_nd[i]*100,2))
#제주
up_down_norm_j = []

for i in range(len(data_nj)-1):
    up_down_norm_j.append(round((data_nj[i+1]-data_nj[i])/data_nj[i]*100,2))
    
up_down_norm_s
```
